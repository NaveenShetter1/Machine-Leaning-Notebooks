```python
# we can remove travel_type_id as this is unique 
# we can remove package id as all values are null
# to_area_id we can remove because all values are null
# from lat, from_long, to_lat, to_long we can remove as to_lat, to_long has no values so no use of from_lat and from_long
```

```python
df_ttif1=df_ttif1.drop(['travel_type_id','package_id','to_area_id','from_lat','from_long','to_lat','to_long'],axis=1)
df_ttif1.head()
```

```python
df_ttif1.isnull().sum()
```

```python
# now we have missing values we need to impute them with correct values by mode 
# here in from_city_id we have more than 10% data is missing
```

```python
df_ttif1.info()
```

```python
# here from_area_id, from_city_id, to_city_id should be object, because they are categorical values

df_ttif1=df_ttif1.astype({'from_area_id':'object','from_city_id':'object','to_city_id':'object'})
df_ttif1.info()
```

```python
df_ttif1.describe(include=['object'])
```

```python
# here top is mode freq means number of times mode occured, so you can impute those values in place of null values
```

```python
df_ttif1.from_area_id=df_ttif1.from_area_id.fillna(393)
df_ttif1.from_city_id=df_ttif1.from_city_id.fillna(15)
df_ttif1.to_city_id=df_ttif1.to_city_id.fillna(32)
```

```python
df_ttif1.isnull().sum()
```

```python
# lets create high, medium and low, zero cancellation

from_area_id_cancellation_perc=df_ttif1[['from_area_id','Car_Cancellation']].groupby('from_area_id').mean()['Car_Cancellation']
from_area_id_cancellation_perc.head(20)
```

```python
# now we need to map respective values
df_ttif1['from_area_id_cancellation_perc']=df_ttif1['from_area_id'].map(from_area_id_cancellation_perc)
df_ttif1.head()
```

```python
df_ttif1[df_ttif1['from_area_id']==571.0].from_area_id_cancellation_perc
```

```python
df_ttif1[df_ttif1['from_area_id']==571.0]
```

```python
# now creating routes
df_ttif1['city_route']=df_ttif1['from_city_id'].astype('int').astype('str')+ ' -> '+ df_ttif1['to_city_id'].astype('int').astype(str)
df_ttif1.head()
```

```python
# city route cancellation %

city_route_cancelation_per=df_ttif1[['city_route','Car_Cancellation']].groupby('city_route').mean()['Car_Cancellation']
df_ttif1['city_route_cancelation_per']=df_ttif1['city_route'].map(city_route_cancelation_per)
df_ttif1.head()
```

```python
print(df_ttif1.columns)
```

```python
df_ttif1['from_area_id_cancellation']=np.where(df_ttif1['from_area_id_cancellation_perc']>=0.67, 'High',
np.where(df_ttif1['from_area_id_cancellation_perc']==0, 'Zero',                         
np.where((df_ttif1['from_area_id_cancellation_perc']>0.33) & (df_ttif1['from_area_id_cancellation_perc']<0.67), 'Medium','Low')))

df_ttif1.head()
```

```python
df_ttif1['from_area_id_cancellation'].unique()
```

```python
df_ttif1['city_route_cancelation']=np.where(df_ttif1['city_route_cancelation_per']>=0.67, 'High',
np.where(df_ttif1['city_route_cancelation_per']<0.33, 'Low', 'Medium',))

df_ttif1['city_route_cancelation']=np.where(df_ttif1['city_route_cancelation_per']==0,'Zero',df_ttif1['city_route_cancelation'])

df_ttif1.head()
```

```python
new_df1=df_ttif1[['from_area_id','from_area_id_cancellation_perc','from_area_id_cancellation','from_city_id','to_city_id','city_route','city_route_cancelation_per','city_route_cancelation']].head()
```

```python
df_ttif1=df_ttif1.drop(['from_area_id','from_area_id_cancellation_perc','from_city_id','to_city_id','city_route','city_route_cancelation_per'],axis=1)
df_ttif1.head()
```
