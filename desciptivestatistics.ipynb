```python
# IQR method to find outliers
```

```python
q1=df1['total_bill'].quantile(0.25)
q3=df1['total_bill'].quantile(0.75)
print('q1: ',q1)
print('q3: ',q3)
iqr=q3-q1

posoutliers=q3+1.5*iqr
negoutliers=q1-1.5*iqr

for i in df1['total_bill']:
  if i>posoutliers or i<negoutliers:
    print(i)
```

```python
df1['TotalBill_zscore']=zscore(df1['total_bill'])

df1.head()
```

```python
'''

The Z-score (also known as standard score) is a statistical measurement that describes a data point's position
relative to the mean of a group of data points, in terms of standard deviations. In other words, it tells you how
many standard deviations a specific data point is away from the mean of the dataset.


Z > 3 or Z < -3: In most cases, Z-scores greater than 3 or less than -3 are considered outliers, because they represent data
points that are significantly distant from the mean.
'''
```

```python
# print z score whole value is > 3 and <-3 they are considered to be outliers
```

```python
df1[df1['TotalBill_zscore']>3]
```

```python
df1[df1['TotalBill_zscore']<-3]
```

```python
for i in df1['TotalBill_zscore']:
  if i>3 or i<-3:
    print(i)
# so here we can say the data is only having outliers in one side means right side so called as right skewed
```

```python
# now above you can compare with outliers which you predicted before by using IQR method
```

```python
# Both IQR (Interquartile Range) and Z-score are commonly used methods for identifying outliers.
# The Z-score method is more appropriate when the data is normally distributed.


# The IQR method is widely used and often considered more robust for detecting outliers,
# especially when the data is not normally distributed or contains non-normal distributions (e.g., skewed data, multimodal data).
```
