```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
# convert categorical column to numerical
```

```python
for i in df.select_dtypes(include=['object']):
  df[i]=df[i].astype('category')
  df[i]=df[i].cat.codes
  df=pd.get_dummies(df,columns=['Gender','Geography'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# df=pd.get_dummies(df,columns=['Gender','Geography'])
```

```python
df.head()
```

```python
# exited is target so 1 means no continueing by user
```

```python
x=df.drop(['Exited'],axis=1)
```

```python
x.head()
```

```python
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
print(x_test.shape)
print(x_test[:5])  # Print first 5 rows
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
x_train = scaler.fit_transform(x_train)  # Fit on training data
x_test = scaler.transform(x_test)  # Only transform test data
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()

svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report, precision_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred)).plot(cmap='Blues')
```

```python
classification_report(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
# so 80% is one class and 20% one class: so oversampling is used to increase minority class counts
```

```python
# SMOTE synthetic minority oversampling technique to remove or reduce class imbalance
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=11)
x_over,y_over=sm.fit_resample(x_train,y_train)
```

```python
x_over.shape
```

```python
y_over.shape
```

```python
y_over.value_counts()
```

```python
(y_over.value_counts()[0])/len(y_over)*100
```

```python
(y_over.value_counts()[1])/len(y_over)*100
```

```python
from sklearn.svm import SVC

svc=SVC(kernel='linear', C=1, probability=True)

svc.fit(x_over,y_over)
y_pred=svc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred)).plot(cmap='Blues')
```

```python
classification_report(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```
