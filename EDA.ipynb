```python
x=df.groupby(by='Class').agg(raisehand=('raisedhands','mean'),
                            Resources=('VisITedResources','mean'),
                            Discussion=('Discussion','mean'),
                            Announcement=('AnnouncementsView','mean'))
print(x)
print('-')
plt.figure(figsize=(12,10))

plt.subplot(1,4,1)
plt.pie(labels=x.index,x=x['raisehand'].values,autopct='%.2f')
plt.title('Class & average Raisehand')

plt.subplot(1,4,2)
plt.pie(labels=x.index,x=x['Resources'].values,autopct='%.2f')
plt.title('Class & average Resource')

plt.subplot(1,4,3)
plt.pie(labels=x.index,x=x['Discussion'].values,autopct='%.2f')
plt.title('Class & average Discussion')

plt.subplot(1,4,4)
plt.pie(labels=x.index,x=x['Announcement'].values,autopct='%.2f')
plt.title('Class & average Announcment')
```

```python
plt.figure(figsize=(7,7))
sns.swarmplot(x=df['Class'],y=df['Discussion'],hue=df['StudentAbsenceDays'])

print('''Students having absence below 7 days are mostly securing higher grades(90-10) and
more number of times participating in the discussion.''')
print('-')
print('''Students having absence above 7 days are mostly securing lower grades(below 69)and
less number of times participating in the discussion.''')
```

```python
plt.figure(figsize=(7,7))
sns.swarmplot(x=df['Class'],y=df['VisITedResources'],hue=df['ParentschoolSatisfaction'])

print('''Parents of those students are satisifed with school who are secured higher grade(above 90)
 and who have visited resources more number of times.''')
print('-')
print('''Parents of those students are not satisifed with school who are secured lower grade(below 69)
 and who have visited resources lesser number of times.''')
```

```python
print(len(df[(df['Discussion']>=df['Discussion'].mean()) & (df['Class']=='H' ) | (df['Class']=='M')].head()))
df[(df['Discussion']>=df['Discussion'].mean()) & (df['Class']=='H' ) | (df['Class']=='M')].head()
```

```python
a=len(df[(df['Topic']=='IT') & (df['Class']=='H')])
b=len(df[(df['Topic']=='IT') & (df['Class']=='M')])
c=len(df[df['Topic']=='IT'])

print((a+b)/c*100)
```

```python
d=[]
e=[]
for i in df['Topic'].value_counts().index:
    a=len(df[(df['Topic']==i) & (df['Class']=='H')])
    b=len(df[(df['Topic']==i) & (df['Class']=='M')])
    c=len(df[df['Topic']==i])
    d.append(i)
    e.append(a+b/c*100)
    print(f'The pecentage students securing between 70 to 100 % marks in {i} is ',round((a+b/c*100),2))
# print(d)
# print(e)
sns.barplot(x=d,y=e)
plt.xticks(rotation=90)
plt.show()
```

```python
d=[]
e=[]
for i in df['Topic'].value_counts().index:
    a=len(df[(df['Topic']==i) & (df['Class']=='H')])
    c=len(df[df['Topic']==i])
    d.append(i)
    e.append(a/c*100)
    print(f'The pecentage students above 90 % marks in {i} is ',round((a/c*100),2))
# print(d)
# print(e)
sns.barplot(x=d,y=e)
plt.xticks(rotation=90)
plt.show()

# More than 50% of the total students of Biology have secured more than 90% marks in exam.
# Only 16% of the toal students of IT have secured more than 90% marks in exam.
```
