```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/mnist_012.csv')
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
len(df.columns)
```

```python
# consider 1 column here as target value
```

```python
xNew=df.drop(['0'],axis=1)
```

```python
xNew.head(3)
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(strategy='median')
x=imputer.fit_transform(xNew)
```

```python
# x=pd.DataFrame(x)
```

```python
# x.isnull().sum()
```

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=100)
pca.fit_transform(x)  # Fit and transform the data
```

```python
print(type(pca))
```

```python
variances = pca.explained_variance_ratio_.round(4)  # ✅ Correct
print(variances)
```

```python
print(type(variances))
```

```python
variances.sum()*100
```

```python
# so it is giving 93% of variance of original data to new df
```

```python
percentages=variances*100
```

```python
percentages
```

```python
var=np.cumsum(np.round(variances,decimals=4))*100
```

```python
var
```

```python
r=np.arange(1,101)
```

```python
r
```

```python
final=[r,variances,percentages,var]
```

```python
final
```

```python
df=pd.DataFrame(final)
```

```python
df1=df.T
```

```python
df1
```

```python
type(df1)
```

```python
columns=['components','variance','percentage','cumulative']
```

```python
df1.columns=columns
```

```python
plt.figure(figsize=(12,6))
pc_values = np.arange(pca.n_components_) + 1
plt.plot(pc_values, pca.explained_variance_ratio_, 'o-', linewidth=1.5, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
```
